```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df1, df2, df3, df4 are already loaded and cleaned
# Define numeric columns
numeric_columns = ['ENG', 'KIS', 'MAT', 'BIO', 'PHY', 'CHE', 'HIS', 'GEO', 'CRE', 'HSC', 'AD', 'AGR', 'COM', 'FRE', 'GER', 'MUS', 'BST']

# Calculate means for each DataFrame
means_df1 = df1[numeric_columns].mean()
means_df2 = df2[numeric_columns].mean()
means_df3 = df3[numeric_columns].mean()
means_df4 = df4[numeric_columns].mean()

# Combine the means into a single DataFrame
means_combined = pd.DataFrame({
    'df1': means_df1,
    'df2': means_df2,
    'df3': means_df3,
    'df4': means_df4
})

# Transpose to have subjects as rows and DataFrames as columns
means_combined = means_combined.T

# Print the DataFrame structure to verify correctness
print(means_combined)

# Plot the line chart
plt.figure(figsize=(12, 6))
for subject in numeric_columns:
    if subject in means_combined.columns:
        plt.plot(means_combined.columns, means_combined.loc[subject], marker='o', label=subject)
    else:
        print(f"Subject '{subject}' not found in means_combined")

plt.title('Mean Scores per Subject Across DataFrames')
plt.xlabel('DataFrame')
plt.ylabel('Mean Score')
plt.legend(title='Subjects')
plt.grid(True)
plt.show()
```
